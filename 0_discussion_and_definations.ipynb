## Choosing a build backend

Tools like **pip** and **build** do not actually convert your sources into a **distribution package** (like a **wheel**); that job is performed by a build backend. The **build backend** determines how your project will specify its configuration, including metadata (information about the project, for example, the name and tags that are displayed on PyPI) and input files. Build backends have different levels of functionality, such as whether they support building **extension modules**, and you should choose one that suits your needs and preferences.

The pyproject.toml tells **build frontend** tools like pip and build which backend to use for your project. Below are some examples for common build backends, but check your backendâ€™s own documentation for more details.

**pip**: The most popular tool for installing Python packages, and the one included with modern versions of Python.

**build**: A tool that helps prepare your Python project to be packaged, but it doesnâ€™t actually make the final installable file itself â€” it asks the build backend to do that. Think of it like a helper that organizes everything before the â€œfactoryâ€ makes the final product.

**distribution package**: The ready-to-install version of your project that others can download and use. Itâ€™s like a box containing all the files, instructions, and info someone needs to install your software.

**wheel**: A specific type of distribution package for Python. You can think of it as a pre-packed box thatâ€™s easy and fast to install â€” Python can just â€œunpack itâ€ without extra work.

**build backend**: The â€œfactoryâ€ that actually creates the distribution package from your project. It reads your projectâ€™s settings, metadata, and code, and produces the final installable file. Different backends have different abilities depending on what kind of project you have.

**extension module**: A piece of Python code that is written in another language, usually C or C++, to make it run faster or do things Python alone canâ€™t. You can still use it like a regular Python module once itâ€™s built.

**build frontend**: A tool you use to tell the build backend to make your package. It doesnâ€™t do the actual building itself â€” itâ€™s more like a â€œproject managerâ€ that organizes the work and gives instructions to the backend (the â€œfactoryâ€). Example:
	â€¢	build is a frontend â€” you run python -m build, and it calls the backend (like Hatchling or Setuptools) to create the wheel or sdist.
	â€¢	pip can also act as a frontend when installing a package: it downloads the source and tells the backend to build it if needed.


### How to choose a Python Build Backend

When packaging Python projects, the **build backend** is the tool that takes your project code and metadata and produces a distribution package (wheel, sdist). Choosing the right backend depends on your projectâ€™s complexity, features, and workflow preferences.

---

#### 1ï¸âƒ£ What a Build Backend Does
- Reads your project metadata (e.g., `[project]` table in `pyproject.toml`)  
- Builds a **distribution package** (wheel, sdist)  
- Handles optional features like **extension modules** (C/C++ code), custom scripts, and dependency resolution  

---

#### 2ï¸âƒ£ Common Build Backends

| Backend | Pros | Cons | Use Case |
|---------|------|------|----------|
| **Hatchling** | Modern, very fast, simple config, supports `[project]` metadata | Smaller ecosystem, fewer plugins than setuptools | Great for new projects with standard Python packages |
| **Setuptools** | Very mature, flexible, huge ecosystem, lots of tutorials | Can be more verbose, older style configuration | Projects with C extensions, legacy code, or complex packaging needs |
| **Flit** | Extremely simple, automatic metadata handling | Limited flexibility for complex builds | Small pure-Python libraries, beginner-friendly |
| **PDM** | Full Python dependency management + packaging | Slightly newer, less widely adopted | Projects where you also want a modern dependency manager |

---

#### 3ï¸âƒ£ How to Choose for Your Use Case

1. **Simplicity:**  
   - For a **pure Python package**, use **Hatchling** or **Flit** for minimal configuration.  

2. **Complexity / C extensions:**  
   - If your project has **extension modules**, custom build steps, or legacy requirements, **Setuptools** is safer.  

3. **Modern workflow:**  
   - If you want **all-in-one tools** (dependencies + packaging), **PDM** is a good choice.  

4. **Community support & tutorials:**  
   - **Setuptools** has the largest support.  
   - **Hatchling** is simpler, modern, and fully compatible with `[project]` metadata.  

---

##### ðŸ’¡ Rule of Thumb
- **New, pure Python library â†’ Hatchling or Flit**  
- **Needs C extensions / complex build â†’ Setuptools**  
- **Want modern all-in-one â†’ PDM**

**Short answer:**

### Cnfiguring Build-Backend

```python
[build-system]
requires = ["hatchling >= 1.26"]
build-backend = "hatchling.build"
```

The **requires** key is a list of packages that are needed to build your package. The frontend should install them automatically when building your package. Frontends usually run builds in isolated environments, so omitting dependencies here may cause build-time errors. This should always include your backendâ€™s package, and might have other build-time dependencies. The minimum version specified in the above code block is the one that introduced support for the new **license metadata**.

The **build-backend key** is the name of the Python object that frontends will use to perform the build.

Both of these values will be provided by the documentation for your build backend, or generated by its command line interface. There should be no need for you to customize these settings.

Additional configuration of the build tool will either be in a tool section of the pyproject.toml, or in a special file defined by the build tool. For example, when using setuptools as your build backend, additional configuration may be added to a setup.py or setup.cfg file, and specifying setuptools.build_meta in your build allows the tools to locate and use these automatically.
